```python
import cProfile
import pstats
from pstats import SortKey

import numpy as np
import xgi
```

```python
p = np.array([[0.3, 0.01], [0.01, 0.3]])
H = xgi.uniform_HSBM(30, 2, p, [15, 15])
Hprime = xgi.uniform_HSBM(50, 2, p, [25, 25])
```

```python
cProfile.run("xgi.simple_cnm(H)", "cnm_n-30.profile")
```

```python
cProfile.run("xgi.simple_cnm(Hprime)", "cnm_n-50.profile")
```

```python
p = pstats.Stats('cnm_n-30.profile')
q = pstats.Stats('cnm_n-50.profile')
```

```python
p.sort_stats(SortKey.CUMULATIVE).print_stats('xgi', 10)
```

```python
q.sort_stats(SortKey.CUMULATIVE).print_stats('xgi', 10)
```
